```python
cat_cols = ['Survived','Pclass','Sex','Embarked']

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    df[col].value_counts(dropna=False).plot(kind='bar', ax=axes[i])
    axes[i].set_title(f'Distribusi {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Jumlah')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns # Import seaborn for enhanced visualization

num_only = df.select_dtypes(include=[np.number])
corr = num_only.corr(numeric_only=True)

plt.figure(figsize=(8, 6)) # Increased figure size for better readability
sns.heatmap(corr, cmap='coolwarm', vmin=-1, vmax=1, annot=True, fmt=".2f") # Use seaborn.heatmap with annotation
plt.title('Heatmap Korelasi (numerik)')
plt.tight_layout()
plt.show()
```

```python
# Import library
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Memuat data
df = pd.read_csv('/content/wbc.csv')

# Memisahkan variabel yang dapat digunakan dan tidak dapat digunakan
unused_var = df[['id']]  # Variabel tidak digunakan (id)
target_var = df[['diagnosis']]  # Variabel target
feature_vars = df.drop(columns=['id', 'diagnosis'])  # Variabel fitur numerik

# Tampilkan informasi untuk verifikasi
print("Variabel yang tidak digunakan (id):")
print(unused_var.head())
print("\nVariabel target (diagnosis):")
print(target_var.head())
print("\nVariabel fitur numerik:")
print(feature_vars.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Memisahkan variabel
target_var = df[['diagnosis']].copy()  # Gunakan .copy() untuk menghindari warning

# Encoding kolom diagnosis
le = LabelEncoder()
target_var['diagnosis'] = le.fit_transform(target_var['diagnosis'])

# Tampilkan hasil encoding
print("Hasil encoding kolom diagnosis:")
print(target_var.head())

# Verifikasi distribusi (opsional)
print("\nDistribusi nilai setelah encoding:")
print(target_var['diagnosis'].value_counts())
```

```python

# Memisahkan variabel
feature_vars = df.drop(columns=['id', 'diagnosis', 'Unnamed: 32'])  # Hapus kolom tidak relevan dan kosong

# Memeriksa dan menangani nilai NaN (jika ada)
print("Jumlah nilai NaN per kolom sebelum standarisasi:")
print(feature_vars.isnull().sum())

# Standarisasi fitur numerik
scaler = StandardScaler()
feature_vars_scaled = scaler.fit_transform(feature_vars)

# Konversi hasil standarisasi kembali ke DataFrame
feature_vars_scaled = pd.DataFrame(feature_vars_scaled, columns=feature_vars.columns)

# Tampilkan hasil standarisasi
print("\nHasil standarisasi fitur numerik:")
print(feature_vars_scaled.head())
```
